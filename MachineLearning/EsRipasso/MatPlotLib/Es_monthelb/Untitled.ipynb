```python
from matplotlib import pyplot
import matplotlib.pyplot as plt
import numpy as np

dictionary = {"data":[], "ora":[], "id_rete":[], "cod_istat":[], "progr_punto":[], "velv":[],
              "dirv":[], "devst":[], "baro":[], "terma":[], "igro":[], "radd":[], 
              "radr":[], "pluv":[]}

file = open("MonthElb_012010_004028900.csv").readlines()

for row in file:

    row = row.split(",")
    if "9999.9" not in (row[9], row[10]) :
        dictionary["data"].append(row[0])
        dictionary["ora"].append(row[1])
        dictionary["id_rete"].append(row[2])
        dictionary["cod_istat"].append(row[3])
        dictionary["progr_punto"].append(row[4])
        dictionary["velv"].append(float(row[5]))
        dictionary["dirv"].append(row[6])
        dictionary["devst"].append(row[7])
        dictionary["baro"].append(row[8])
        dictionary["terma"].append(float(row[9]))
        dictionary["igro"].append(float(row[10]))
        dictionary["radd"].append(row[11])
        dictionary["radr"].append(row[12])
        dictionary["pluv"].append(row[13])


fig, ax = plt.subplots(figsize=(8,8))

ax.scatter(dictionary["igro"], dictionary["terma"], alpha=0.5)
ax.set_xlabel("Umitidà")
ax.set_ylabel("Temperatura dell'aria")

plt.title("Grafico umidità - Temperatura del'aria")
plt.show()
```
