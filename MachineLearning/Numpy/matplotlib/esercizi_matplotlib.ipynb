### linestyle (o ls) definisce lo stile del tratto

* '-' o 'solid' → Linea continua  
* '--' o 'dashed' → Linea tratteggiata  
* '-.' o 'dash_dot' → Alternanza punto-trattino  
* ':' o 'dotted' → Linea punteggiata  
* '' o ' ' o 'None' → Nessuna linea  

```python
x = np.linspace(0.,10.,50)
y = np.sin(x)
pl.plot(x,y, linestyle='--')
pl.show()
```

```python
x = np.linspace(0.,10.,50)
y = np.sin(x)
pl.plot(x,y, linewidth = 5)
pl.show()
```

### Stiple Marker (punto ciascun valore)

* '.' → Punto '+' → Croce
* ',' → Pixel 'x' → X
* 'o' → Cerchio 'D' → Rombo
* 's' → Quadrato '*' → Asterisco
* 'v' '^' '<' '>' → Triangoli (orientati a seconda del carattere utilizzato)
* '' o ' ' o 'None' → Nessun punto

```python
x = np.linspace(0.,10.,50)
y = np.sin(x)
pl.plot(x,y, marker = 'o')
pl.show()
```

```python
x = np.linspace(0.,10.,50)
y = np.sin(x)
pl.plot(x,y, marker = 'o', markersize = 9)
pl.show()
```

### Colori Linea

* 'b' o 'blue' → Blu
* 'g' o 'green' → Verde
* 'r' o 'red' → Rosso
* 'm' o 'magenta'→ Magenta
* 'k' o 'black' → Nero
* 'y' o 'yellow' → Giallo
* 'c' o 'cyan' → Ciano
* Altri colori → codice esadecimale (es. 'FF5C4A')

```python
import csv
d={}
dataFile = open("marziani.csv","r")
dataReader = csv.reader(dataFile,delimiter=",")
for i,row in enumerate(dataReader):
    if i==0:
        keys=row
    else:
        for k,e in enumerate(row):
            if i!=1:
                d[keys[k]].append(e)
            else:
                d[keys[k]]=[e]

                

fig, ax = pl.subplots(figsize=(50,50))
ax.scatter(d["altezza"], d["peso"], c="red", s=100, marker=".")
ax.set_xlabel("peso", fontsize=20)
ax.set_ylabel("altezza", fontsize=20)
pl.show()
```

```python
from cProfile import label
import csv

import numpy

with open("marziani.csv", newline="", encoding="ISO-8859-1") as filecsv:
    lettore = csv.reader(filecsv,delimiter=",")
    lista = []

    for k,riga in enumerate (lettore):
        
        if k!=0:
            lista.append(riga)
            




dati_marziani = {"tipo": [], "colore": [], "n_arti": [], "peso": [], "altezza": [], "larghezza": []}

for k in lista:
    if k[0] != '' and k[1] != '' and k[2] != '' and k[3] != '' and k[4] != '' and k[5] != '':
        dati_marziani["tipo"].append(k[0])
        dati_marziani["colore"].append(k[1])
        dati_marziani["n_arti"].append(float(k[2]))
        dati_marziani["peso"].append(float(k[3]))
        dati_marziani["altezza"].append(float(k[4]))
        dati_marziani["larghezza"].append(float(k[5]))

for k in dati_marziani:
    dati_marziani[k] = numpy.array(dati_marziani[k])



fig, ax = pl.subplots(figsize=(10,10))
ax.scatter(dati_marziani["peso"][dati_marziani["tipo"] == "Robby"], 
dati_marziani["larghezza"][dati_marziani["tipo"] == "Robby"],
label="Robby",c="red",alpha=0.5 ,s=100, marker=".")

ax.scatter(dati_marziani["peso"][dati_marziani["tipo"] == "Simmy"], 
dati_marziani["larghezza"][dati_marziani["tipo"] == "Simmy"],
label="Simmy",c="green",alpha=0.5 ,s=100, marker=".")

ax.set_xlabel("peso", fontsize=20)
ax.set_ylabel("altezza", fontsize=20)

ax.legend()
pl.show()
```
