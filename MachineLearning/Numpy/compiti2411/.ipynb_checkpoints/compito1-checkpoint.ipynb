```python
'''
Partendo dal file classe.txt creare dei turni di interrogazione
'''
import random
import numpy as np

def main():
    file = open("./Classe.txt")
    classList = []
    nTurn = 1
    
    for row in file.readlines():
        classList.append(row[:-1])
    length = len(classList)
    
    studentiTurni = input("Inserisci n studenti per turno: ")
    
    while classList:
        if len(classList) < int(studentiTurni): 
            turn = np.random.choice(classList, size = len(classList), replace = False) #creazione ultimo turno
        else:
            turn = np.random.choice(classList, size = int(studentiTurni), replace = False)
            
        #rimozione nomi estratti
        for name in turn:
            classList.remove(name)
        
        print(f'Turno {nTurn}: {turn}')
        nTurn+=1
        
    print('\n-----------------------\n')
    
    
    heightArray = np.random.normal()  #np.random.choice(180, length, 30)
    print(heightArray)
        
    print('\n-----------------------\n')
    
    misurationsArray = np.random.normal(19, 5, 1000)
    misurationsArray.shape = (10,100)
    print(misurationsArray)

main()
```
