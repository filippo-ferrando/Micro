```python
import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as pl

i = np.concatenate((pl.imread('./foto1.png'), pl.imread('./foto2.png')), axis=1)
k = np.concatenate((pl.imread('./foto3.png'), pl.imread('./foto4.png')), axis=1)
collage = np.concatenate((i,k), axis=0)

fig,ax = pl.subplots(figsize=(10,10))
ax.imshow(collage)
ax.set_title('Collage', fontsize=20)
ax.axis('off')
pl.savefig('./Collage.png', format='png')
pl.show()
```
