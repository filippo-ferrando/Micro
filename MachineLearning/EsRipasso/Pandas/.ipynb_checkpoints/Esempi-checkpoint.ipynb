# Apertura json

```python
asteroidi_comete = pd.read_json("neowise.json")
asteroidi_comete.head()
```

```python
asteroidi_comete.tail(3) #ultimi 3 dati
```

```python
asteroidi_comete_ridotto = asteroidi_comete[["designation", "discovery_date", "h_mag", "period_yr"]]
asteroidi_comete_ridotto.head()
```

```python
#filtraggio dati singolo
asteroidi_comete_ridotto[asteroidi_comete_ridotto["period_yr"] < 1.0]
```

```python
#filtraggio dati doppio
asteroidi_comete_ridotto[(asteroidi_comete_ridotto["period_yr"] < 1.0)
                         & (asteroidi_comete_ridotto["discovery_date"] > '2014-01-01')]
```

```python
#aggiungere una colonna tramite assegnamento
asteroidi_comete["period_day"] = asteroidi_comete['period_yr']*365
asteroidi_comete.head()
```

```python
asteroidi_comete_rinominato = asteroidi_comete.rename(
    columns={"designation":"nome",
             "discovery_date":"data_scoperta"}
)

asteroidi_comete_rinominato.head()
```

```python
#Rimpiazzare i dati nelle righe
asteroidi_comete_rinominato["pha"].replace('Y','SI',inplace=True)
asteroidi_comete_rinominato["pha"].replace('N','NO',inplace=True)
asteroidi_comete_rinominato.head()
```

```python
np.set_printoptions(suppress = True) #per eliminare la notazione scientifica
asteroidi_comete_rinominato['period_yr'].values()
```

# Marziani

```python
marziani = pd.read_csv('marziani.csv')
marziani
```

```python
fig, ax = plt.subplots(figsize=(8,8))
marziani[marziani["specie"]=="Robby"].plot.scatter(x="peso", y="larghezza", c="g", label="Robby", ax=ax)
marziani[marziani["specie"]=="Simmy"].plot.scatter(x="peso", y="larghezza", c="r", label="Simmy", ax=ax)

ax.set_xlabel("peso", fontsize=16)
ax.set_ylabel("larghezza", fontsize=16)
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)
ax.set_title("Le caratteristiche dei marziani", fontsize=20)
ax.grid(True)
ax.legend(fontsize=16)
plt.show()
```
