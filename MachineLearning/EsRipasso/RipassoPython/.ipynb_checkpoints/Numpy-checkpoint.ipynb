# Creare ndarray da liste, tuple es intervalli

```python
import numpy as np

a1 = np.array((1, 2, 3)) #array da tupla
a2 = np.array([1.6, -1, 3], np.uint8) #array da lista con specifica di tipo
a3 = np.arange(10, dtype = float) #array da range [0...STOP] con passo 1 e specifica di tipo
a4 = np.arange(4, 10, 2) #array da range [START...STOP] con passo 2
a5 = np.linspace(0, 10, 5) #array di 5 numeri equalmente distribuiti in [START...STOP]

print("a1:", a1)
print("a2:", a2)
print("a3:", a3)
print("a4:", a4)
print("a5:", a5)
```

```python
'''
Partendo dal file classe.txt creare dei turni di interrogazione
'''
print('-------------ES 1-------------\n')
import random
import numpy as np

def main():
    file = open(".\Classe.txt")
    classList = []
    nTurn = 1
    
    for row in file.readlines():
        classList.append(row[:-1])
    length = len(classList)
    
    sizeTurno = int(input("Inserisci il numero di persone per turno: "))
    
    while classList:
        if len(classList) < sizeTurno: 
            turn = np.random.choice(classList, size = len(classList), replace = False) #creazione ultimo turno
        else:
            turn = np.random.choice(classList, size = sizeTurno, replace = False)
            
        #rimozione nomi estratti
        for name in turn:
            classList.remove(name)
        
        print(f'Turno {nTurn}: {turn}')
        nTurn+=1
        
    print('\n------------ES 2------------\n')
    
    heightArray = np.random.choice(180, length, 30)
    print(heightArray)
        
    print('\n------------ES 3------------\n')
    
    misurationsArray = np.random.normal(19, 5, 1000)
    misurationsArray.shape = (10,100)
    print(misurationsArray)
    
if __name__ == ("__main__"):
    main()
```

```python
#%%timeit
import numpy as np
nClassi = 10
nAlunni = 10
nGiorni = 365

giorni = np.random.randint(0,nGiorni, size=(nClassi,nAlunni))
print(giorni, '\n')
classe = 0
for row in giorni:
    classe += 1
    ricorrenze = np.bincount(row)
    for value in ricorrenze:
        if value > 1:
            print(f'Classe: {classe} --> almeno due persone nate nello stesso giorno')
            break

'''
cnt = 0
for row in giorni:
    for elementi in row:
        if np.sum(element == row) > 1:
            cnt += 1
            break
'''

print('\n\nSeconda Soluzione\n')
cnt = 0
for row in giorni:
    valori, freqAss = np.unique(row, return_counts = True)
    print(dict(zip(valori, freqAss)))
    if np.count_nonzero(freqAss >= 2) == True: # || >0
        print(np.count_nonzero(freqAss >= 2))
        cnt += 1
print(f'Numero di classi con due alunni nati lo stesso giorno: {cnt}')
```

```python
import numpy as np

array = np.array([0,2,1,1,3,4])
arrayDx = np.roll(array, 1)
arraySx = np.roll(array, -1)
print(array[np.logical_and(array > arrayDx, array > arraySx)], '\n')

#Con estremi esclusi:
array = np.array([0,2,1,1,3,4])
arrayDx = np.roll(array, 1)
arraySx = np.roll(array, -1)
mat = np.array([array, arrayDx, arraySx]).reshape(3,-1)
#mat = np.stack((array, arrayDx, arraySx), axis=0)
print(mat, '\n')
massimo = np.argmax(mat[:,1:-1], axis = 0) #indice del massimo rispetto all'asse 0
print(massimo)
print(f'I massimi locali si trovano nella posizione: {(np.argwhere(massimo==0)+1).flatten()}, i loro valori sono: {array[(np.argwhere(massimo==0)+1).flatten()]}')
#print(array[(np.argwhere(massimo==0)+1).flatten()]) #argwhere trova gli indici della matrice in cui vale quella condizione, flatten ritrasforma la matrice in array
```

```python
import numpy as np

nNumeri = 30000
seq = [1,2,3]
array = np.random.randint(10, size = nNumeri)
print(array)

for i in range(len(array)-len(seq)+1):
    #print(array[i:i+len(seq)] == seq)
    if np.all((array[i:i+len(seq)] == seq))==True: #la all controlla se tutta quella riga è True
        print(i)
        
#Seconda soluzione
mat = np.array([array[i:len(seq)+i] for i in range(len(array)-len(seq)+1)])
print('\n',mat,)
#print(sequenza == mat)
print(np.argwhere(np.all(seq == mat, axis=1)).flatten())
```
