```python
'''
13) Nel file MonthElb_012010_004028900.csv presente in classroom ESERCIZI, 
sono contenute misure meteorologiche dell'anno 2010 con cadenza oraria, 
per una stazione meteo italiana. Le colonne del file csv sono le seguenti:
1. DATA 2. ORA: espressa nel sistema UTC 3. ID_RETE_MONIT: Identificativo della Rete di Monitoraggio 
4. CODICE_ISTAT_COMUNE: Identificativo della stazione secondo codifica ISTAT 
5. PROGR_PUNTO_COM: Numerazione progressiva delle stazioni con medesimo codice ISTAT
6. VELV: Velocità del vento 7. DIRV: Direzione del vento 8. DEVST: Deviazione standard della velocità trasversale 
9. BARO: Pressione atmosferica 10. TERMA: Temperatura dell'aria 11. IGRO: Umidità relativa 
12. RADD: Radiazione solare diretta 13. RADR: Radiazione solare riflessa 
14. PLUV: Precipitazione In caso di valori mancanti è riportato il numero 9999.9. Leggi il file csv, 
crea: 1. un grafico a dispersione avente come ascissa l’umidità relativa e come ordinata la temperatura dell’aria. 
2. una figura contenente 5 x 5 sotto-assi che rappresentino i 25 possibili diagrammi a dispersione e i 5 possibili istogrammi,
considerando solo velocità del vento, pressione atmosferica, temperatura dell'aria, umidità relativa, precipitazione.
'''

import csv
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy

with open("MonthElb_012010_004028900.csv", newline="", encoding="ISO-8859-1") as filecsv:
    lettore = csv.reader(filecsv,delimiter=",")
    lista = []

    for riga in lettore:
        lista.append(riga)

#lista.pop(0)


dati = {"data": [], "ora": [], "idReteMonit": [], "codiceIstatComune": [], "progrPuntoCom": [], "velv": [], "dirv": [], "devSt": [], "baro": [], "terma": [], "igro": [], "radd": [], "radr": [], "pluv": []}

for k in lista:
    if k[0] != '' and k[1] != '' and k[2] != '' and k[3] != '' and k[4] != '' and k[5] != '' and k[6] !='' and k[7] !='' and k[8] !='' and k[9] !='' and k[10] !='' and k[11] !='' and k[12] !='' and k[13] !='9999.9':
        dati["data"].append(k[0])
        dati["ora"].append(k[1])
        dati["idReteMonit"].append(float(k[2]))
        dati["codiceIstatComune"].append(float(k[3]))
        dati["progrPuntoCom"].append(float(k[4]))
        dati["velv"].append(float(k[5]))
        dati["dirv"].append(float(k[6]))
        dati["devSt"].append(float(k[7]))
        dati["baro"].append(float(k[8]))
        dati["terma"].append(float(k[9]))
        dati["igro"].append(float(k[10]))
        dati["radd"].append(float(k[11]))
        dati["radr"].append(float(k[12]))
        dati["pluv"].append(float(k[13]))

for k in dati:
    dati[k] = numpy.array(dati[k])

#print(dati)
```

```python
# plt.plot(dati["igro"],dati["terma"])
# plt.scatter(dati["igro"], dati["terma"])
# plt.show()


caratteristiche = ["igro",'terma']

n = len(caratteristiche)
fig, axs = plt.subplots(n, n, figsize=(16,16), constrained_layout=True)
for j, dato_y in enumerate(caratteristiche):
    for i, dato_x in enumerate(caratteristiche):
        axs[j,i].scatter(dati[dato_x], 
                    dati[dato_y], c="g", alpha = 0.2)
        axs[j,i].set_xlabel(dato_x, fontsize=14)
        axs[j,i].set_ylabel(dato_y, fontsize=14)
    
fig.suptitle('esercizio 2', fontsize=25)
plt.show()
```
