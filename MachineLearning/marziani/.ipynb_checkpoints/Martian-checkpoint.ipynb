```python
import pprint
```

```python
file = open("marziani.csv").readlines()
marz_dict = {}
labels = []
```

```python
for label in file[0].split(","):
    marz_dict[label] = []
    labels.append(label)
```

```python
for row in file[1:]:
    if not (" " in row.split(",") or "" in row.split(",")):  #non considero righe con campi vuoti o nulli    
        for i, campo in enumerate(row[:-1].split(",")):
                if campo.isdigit():
                    marz_dict[labels[i]].append(int(campo))

                elif all(c.isdigit() or c == "." for c in campo):
                    marz_dict[labels[i]].append(float(campo))

                else:
                    marz_dict[labels[i]].append(campo)
```

```python
pprint.pprint(marz_dict)
```
