```python
import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-10, 10, size=50)
y = np.random.uniform(-10, 10, size=50)
colors = np.random.uniform(0,1, size=(50,3))
dims = np.random.uniform(50,150,size=50)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(x,y,c=colors, s=dims, marker="v")
ax.set_xlabel('x', fontsize=30)
ax.set_ylabel('y', fontsize=20)
ax.tick_params(axis='both', labelsize = 16)
ax.grid(True)
plt.show()
```

```python
from cProfile import label
import csv
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy

with open("marziani.csv", newline="", encoding="ISO-8859-1") as filecsv:
    lettore = csv.reader(filecsv,delimiter=",")
    lista = []

    for riga in lettore:
        lista.append(riga)

lista.pop(0)


dati_marziani = {"tipo": [], "colore": [], "n_arti": [], "peso": [], "altezza": [], "larghezza": []}

for k in lista:
    if k[0] != '' and k[1] != '' and k[2] != '' and k[3] != '' and k[4] != '' and k[5] != '':
        dati_marziani["tipo"].append(k[0])
        dati_marziani["colore"].append(k[1])
        dati_marziani["n_arti"].append(float(k[2]))
        dati_marziani["peso"].append(float(k[3]))
        dati_marziani["altezza"].append(float(k[4]))
        dati_marziani["larghezza"].append(float(k[5]))

for k in dati_marziani:
    dati_marziani[k] = numpy.array(dati_marziani[k])



fig, ax = plt.subplots()
ax.scatter(dati_marziani["peso"][dati_marziani["tipo"] == "Robby"], 
dati_marziani["larghezza"][dati_marziani["tipo"] == "Robby"],
label="Robby")

ax.scatter(dati_marziani["peso"][dati_marziani["tipo"] == "Simmy"], 
dati_marziani["larghezza"][dati_marziani["tipo"] == "Simmy"],
label="Simmy")

plt.xlabel("PESO", family='serif', color='r', weight='normal', size = 16, labelpad = 6)
plt.ylabel("LARGHEZZA", family='serif', color='b', weight='normal', size = 16, labelpad = 6)

ax.legend()
plt.show()
```
