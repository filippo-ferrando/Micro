Nell' esercizio delle valutazioni:
- Visualizza quali sono i voti medi, massimi e minimi per le due prove di verifica
- Considerando l'andamento complessivo della classe e usando i percentili, puoi affermare che l'alunno Di Donato ha fatto una prova migliore nella verifica 1 o oppure nella verifica 2) Motiva la risposta
- E' stato più bravo Armadei nella prima verifica o De Luca nella seoconda ? Motiva la tua risposta utilizzando i percentili
- Inoltre, supponi che l'insegnante abbia deciso di assegnare agli assenti il voto medio preso dagli studenti in quella verifica. Indicare a quali studenti conviene questa scelta(la loro media migliora) e a chi no.

```python
import pandas as pd
voti=pd.read_csv("registro_voti.csv")
voti["Voto medio"]=voti.mean(axis=1)
'''
minVal1=voti["Valutazione1"].min()
masVal1=voti["Valutazione1"].max()
medVal1=voti["Valutazione1"].mean()
minVal2=voti["Valutazione2"].min()
masVal2=voti["Valutazione2"].max()
medVal2=voti["Valutazione2"].mean()
print(f"Minimo Valutazione1: {minVal1}")
print(f"Massimo Valutazione1: {masVal1}")
print(f"Media Valutazione1: {medVal1}")
print(f"Minimo Valutazione2: {minVal2}")
print(f"Massimo Valutazione2: {masVal2}")
print(f"Media Valutazione2: {medVal2}")
'''
print("Informazioni sulla Valutazione1\n")
print(voti["Valutazione1"].describe())
print("\nInformazioni sulla Valutazione2\n")
print(voti["Valutazione2"].describe())
```

```python
print(voti[voti["Cognome"]=="di donato"])
```

Di Donato ha fatto meglio nella seconda valutazione perchè la classe in percentuale ha avuto più difficoltà a svolgere la verifica

```python
print(voti[voti["Cognome"]=="amadei"])
print(voti[voti["Cognome"]=="de luca"])
```

Amadei nella prima e nella seconda valutazione cade fra i 50 e i 75 percentili.
De luca nella prima cade fra i 50 e i 75 percentili mentre nella seconda ricade oltre il 75 percentili
De Luca > amadei

```python
temp=voti["Cognome"][voti["Valutazione1"].isnull()]
#print(temp)
for element in temp:
    mediaNuova=(voti["Valutazione2"][voti["Cognome"]==element]+voti["Valutazione1"].mean())/2
    mediaVecchia=(voti["Voto medio"][voti["Cognome"]==element])
    if float(mediaNuova) > float(mediaVecchia):
        print(f"A {element} conviene")
    else:
        print(f"A {element} non conviene")  
        
temp=voti["Cognome"][voti["Valutazione2"].isnull()]

for element in temp:
    mediaNuova=(voti["Valutazione1"][voti["Cognome"]==element]+voti["Valutazione2"].mean())/2
    mediaVecchia=(voti["Voto medio"][voti["Cognome"]==element])
    if float(mediaNuova) > float(mediaVecchia):
        print(f"A {element} conviene")
    else:
        print(f"A {element} non conviene")  
```
