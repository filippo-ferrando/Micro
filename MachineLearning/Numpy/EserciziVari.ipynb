### Usando la funzione numpy.random.normal(mean, dev, shape) generare un ndarray che contenga la stima dell'altezza degli alunni della classe. 
### L'ndarray deve avere lunghezza pari al numero di compagni di classe. I valori mean e dev devono essere stimati in modo tale da simulare la distribuzione dell'altezza degli alunni in cm nel modo più realistico possibile. (chiedete l'altezza a tutti i compagni)

```python
import numpy as np

altezze = np.random.uniform(1.50, 2, size=(17))
altezze = np.around(altezze, decimals=2)
#print(altezze)

media = altezze.mean()
deviazione = altezze.std()
#print(media, deviazione)

a = np.random.normal(media, deviazione, (1,17))
print(a)

print("\n")

altezze=np.random.normal(175,15,17)
print(np.around(altezze, decimals=2))  #(media, deviazione, quantità)

altezze= np.random.uniform(1.50, 2, size=(17))
print(np.around(altezze, decimals=2))
```

## Consideriamo mille classi da 25 alunni ciascuna: per ognuna di esse simuliamo il giorno di nascita di ciascun alunno estraendo un numero a caso tra 0 e 364 inclusi (i giorni dell'anno non bisestile sono 365): calcolare in quante delle mille classi ci sono almeno due alunni nati lo stesso giorno. 

```python
#%%timeit
a = np.random.randint(0,364,(1000,25))
#k = 0
cC = 0
for row in a:
    #k += 1
    u, c = np.unique(row, return_counts=True)
    dup = u[c > 1]
    if dup.size != 0:
        cC += 1
        #print(f"classe {k} ripetizione il/i giorno/i --> {dup}")
print(f"classi con ripetizione --> {cC}")
```

L'aeroporto JFK di New York vede transitare ogni giorno circa 200000 passeggeri. Ad ogni passeggero è misurata la temperatura. L'array temperature contiene tutte le temperature dei 200000 di passeggeri transitati il giorno 1 gennaio 2020, mentre l'array identificativi contiene la lista dei numeri identificativi della carta di viaggio per ciascun. passeggero. Trovare:

• il numero totale di passeggeri con temperatura maggiore di 37.5°C

• tutti gli identificativi dei passeggeri con temperatura maggiore di 37.5°C

Iniziamo creandoci un array di temperature, ricordandoci che la temperatura media di una persona sana è di 36.5°C e che i campioni avranno una distribuzione gaussiana.

```python
import uuid

passeggeri = 20
array_id = []
temp_array = np.random.normal(36.5, 0.9, passeggeri)

print(temp_array)

print("\n")

array_id = np.array([uuid.uuid4() for _ in range(passeggeri)])
print(array_id)

print("\n")

full_array = dict(zip(array_id, temp_array))
print(full_array)

print("\n")

print(f"persone maggiore di 37.5 = {np.count_nonzero(temp_array > 37.5)}")

print("\n")

print(array_id[temp_array > 37.5])
```

## trova il massimo locale

```python
import numpy as np

a = np.array([0,2,1,1,3,2])
print(a[np.logical_and(a > np.roll(a, 1), a > np.roll(a, -1))])

#estremi esclusi
array= np.array([0,2,1,1,3,2])
arrayDx= np.roll(array,1)[1:-1]   #trasla il vettore di 1 a destra
arraySx= np.roll(array,-1)[1:-1]  #trasla il vettore di (-1 a destra) ovvero 1 a sinistra
a= array[1:-1]
print(a[np.logical_and(a> arrayDx, a>arraySx)])
```

```python
import numpy as np

nNumeri = 3000
seq = [1,2,3]
a = np.random.randint(10, size=nNumeri)
print(a)

for i in range(len(a)-len(seq)+1):
    if np.all((a[i:i+len(seq)] == seq)) == True:
        print(i)
        
print('a', a)
mat = np.array([a[i:len(seq)+i] for i in range(len(a)-len(seq)+1)])
print(mat)
print("\n")
print(np.argwhere(np.all(seq == mat, axis=1)).flatten())
```
