```python
#12 Utilizza numpy per creare una matrice 100 × 100 di numeri interi compresi tra 0 e 255.
import pandas as pd
import numpy as np

numeri = np.random.randint(0,256,(100,100))
#numeri = np.random.normal(0,256,(100,100))
numeri
```

```python
#Crea una figura con due grafici, cioè due sottografici: 
#›› il primo rappresenta la matrice con la mappa di colore scelta tra 4 possibili; 
#›› il secondo rappresenta la distribuzione dei valori degli elementi della matrice come un istogramma.
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1,2, figsize=(5,5))
axs[0].imshow(numeri)
axs[1].hist(numeri)

plt.show()
```

```python
#7 Scarica il file neowise.jso. Esso contiene i dati di asteroidi e comete. 
#Crea un notebook in cui: 
#›› è creato un dataframe contenente i dati; 
#›› è calcolata la media del periodo orbitale (colonna period_yr)
#    per ogni categoria di classe orbitale (la colonna orbit_class); 
#›› utilizzando i metodi grafici dei dataframe, è tracciato un grafico a dispersione 
#    che rappresenti i valori della colonna moid_au rispetto a quelli di period_yr.

asteroidi = pd.read_json("neowise.json")
asteroidi
```

```python
#›› è calcolata la media del periodo orbitale (colonna period_yr)
#    per ogni categoria di classe orbitale (la colonna orbit_class);

mediaPeriodoOrbitale = asteroidi.groupby("orbit_class")["period_yr"].mean()
mediaPeriodoOrbitale
```

```python
#›› utilizzando i metodi grafici dei dataframe, è tracciato un grafico a dispersione 
#    che rappresenti i valori della colonna moid_au rispetto a quelli di period_yr.
import matplotlib.pyplot as plt
asteroidi.plot(x="period_yr",y="moid_au", kind = "scatter")
plt.show()
```
