# Distibuzione dei miei voti attraverso un istogramma

```python
import numpy as np

from matplotlib import pyplot
import matplotlib.pyplot as pl

votes = np.array([8.5, 9, 9, 5.75, 9.25, 8, 8, 10, 7.5, 8, 9, 8.5, 10, 6.5, 8, 7.5, 
                  8.5, 8.5, 7.25, 7.25, 8, 8.25, 9, 9.5, 8, 9.5, 8.25, 9.5, 8, 8.5,
                  8.75, 8, 8.5, 8.75, 8.5, 8.5])

fig,ax = pl.subplots(figsize=(8,8))
ax.hist(votes, bins=10)
ax.set_xlabel("voto",fontsize=16)
ax.set_ylabel("frequenza assoluta", fontsize=16)
ax.tick_params(axis="x", labelsize=16)
ax.tick_params(axis="x", labelsize=16)
ax.set_title("La distribuzione dei miei voti", fontsize=20)

pl.show()
```
